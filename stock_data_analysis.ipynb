```python
import pandas as pd

df = pd.read_csv("combined_stock_data.csv")
df
```

What is the average value of each numerical column?

```python
average_values = df.mean()
print("Average value of each numerical column:")
print(average_values)
```

What is the highest closing price (Close) in the dataset?

```python
highest_closing_price = df['Close'].max()
print("\nHighest closing price (Close) in the dataset:", highest_closing_price)
```

How many unique companies are included in the dataset (Company Name)?

```python
unique_companies = df['Company Name'].nunique()
print("\nNumber of unique companies included in the dataset:", unique_companies)
```

What is the range of values for the Volume column?

```python
volume_range = df['Volume'].max() - df['Volume'].min()
print("\nRange of values for the Volume column:", volume_range)
```

How many unique industries are present in the dataset (Industry)?

```python
unique_industries = df['Industry'].nunique()
print("\nNumber of unique industries present in the dataset:", unique_industries)
```

What is the average turnover for each industry?

```python
average_turnover_by_industry = df.groupby('Industry')['Turnover'].mean()
print("\nAverage turnover for each industry:")
print(average_turnover_by_industry)
```

What is the most traded series (Series) in terms of volume?

```python
most_traded_series = df.groupby('Series')['Volume'].sum().idxmax()
print("\nMost traded series (Series) in terms of volume:", most_traded_series)
```

How many unique ISIN codes are there in the dataset?

Time-Series Analysis:
How has the closing price (Close) of a specific series changed over time?

```python
import pandas as pd
import matplotlib.pyplot as plt

specific_series_data = df[df['Symbol'] == 'DRREDDY']

plt.figure(figsize=(10, 6))
plt.plot(specific_series_data['Date'], specific_series_data['Close'], color='blue', marker='o', linestyle='-')
plt.title('Closing Price Over Time for DRREDDY')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

Compare the closing prices of multiple series on the same plot.

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Symbol': ['DRREDDY', 'DRREDDY', 'HEROMOTOCO', 'HEROMOTOCO'],
    'Date': ['2024-02-14', '2024-02-15', '2024-02-14', '2024-02-15'],
    'Close': [100, 110, 50, 55]
}

df = pd.DataFrame(data)

df['Date'] = pd.to_datetime(df['Date'])

multiple_series_data = df[df['Symbol'].isin(['DRREDDY', 'HEROMOTOCO'])]

plt.figure(figsize=(10, 6))
for series in multiple_series_data['Symbol'].unique():
    series_data = multiple_series_data[multiple_series_data['Symbol'] == series]
    plt.plot(series_data['Date'], series_data['Close'], label=series)

plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Closing Prices Over Time for Multiple Series')
plt.legend()
plt.show()
```

Create a bar chart to compare the turnover for different industries.

```python
import matplotlib.pyplot as plt

import pandas as pd
data = {
    'Industry': ['Finance', 'Technology', 'Healthcare', 'Retail'],
    'Turnover': [1000000, 1500000, 800000, 1200000]
}
df = pd.DataFrame(data)

average_turnover_by_industry = df.groupby('Industry')['Turnover'].mean().sort_values(ascending=False)

average_turnover_by_industry.plot(kind='bar')
plt.xlabel('Industry')
plt.ylabel('Average Turnover')
plt.title('Average Turnover by Industry')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

Visualize the VWAP (Volume Weighted Average Price) over time.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03'],
    'VWAP': [100, 110, 105]
}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])

plt.plot(df['Date'], df['VWAP'])
plt.xlabel('Date')
plt.ylabel('VWAP')
plt.title('VWAP Over Time')
plt.show()
```
