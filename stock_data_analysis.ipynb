Statistical Tests:
Perform a t-test to compare the means of two different series.


```python
import pandas as pd
from scipy import stats

tatasteel_prices = pd.Series([100, 105, 110, 115, 120])
reliance_prices = pd.Series([95, 100, 105, 110, 115])

t_statistic, p_value = stats.ttest_ind(tatasteel_prices, reliance_prices)
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

Explore the correlation significance between 'High' and 'Low' prices.

```python
import pandas as pd
from scipy import stats
import numpy as np

file_path = '/content/combined_stock_data.csv'

try:
    df = pd.read_csv(file_path)

    print("Columns in DataFrame:", df.columns)

    if 'High' in df.columns and 'Low' in df.columns:
        # Handle missing values
        df.dropna(subset=['High', 'Low'], inplace=True)  # Drop rows with NaN values in 'High' or 'Low' columns

        # Handle infinite values
        df.replace([np.inf, -np.inf], np.nan, inplace=True)  # Replace infinite values with NaN
        df.dropna(subset=['High', 'Low'], inplace=True)  # Drop rows with NaN values after replacing infinite values

        # Calculate correlation
        high_low_corr, p_value = stats.pearsonr(df['High'], df['Low'])
        print("Correlation coefficient:", high_low_corr)
        print("P-value:", p_value)
    else:
        print("The 'High' and/or 'Low' columns are not present in the DataFrame.")

except FileNotFoundError:
    print(f"File '{file_path}' not found. Please provide the correct file path.")
```

Market Trends:
Identify periods of market trends using technical analysis.


```python
short_window = 12
long_window = 26
df['ShortEMA'] = df['Close'].ewm(span=short_window, adjust=False).mean()
df['LongEMA'] = df['Close'].ewm(span=long_window, adjust=False).mean()
df['MACD'] = df['ShortEMA'] - df['LongEMA']

plt.plot(df['Date'], df['MACD'], label='MACD')
plt.xlabel('Date')
plt.ylabel('MACD')
plt.title('Moving Average Convergence Divergence (MACD)')
plt.legend()
plt.show()
```
